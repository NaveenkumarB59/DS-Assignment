```python
import pandas as pd
import numpy as np, cv2
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import style
df=pd.read_csv("/content/sales_data_with_discounts.csv")
df
```

```python
df.describe()
```

```python

# Data Visulaization
plt.hist(df['Date'],bins=300,color='red')
plt.show()
```

```python
plt.hist(df['Volume'],bins=70,color='black')
plt.show()
```

```python
plt.hist(df['Avg Price'],bins=50)
plt.show()
```

```python
plt.hist(df['Total Sales Value'],bins=30,color='yellow')
plt.show()
```

```python
plt.hist(df['Discount Rate (%)'],bins=90)
plt.show()
```

```python
random_color = '#%06x' % np.random.randint(0, 0xFFFFFF)
plt.hist(df['Discount Amount'],bins=40,color=random_color,edgecolor='black')
plt.show()
```

```python
random_color = '#%06x' % np.random.randint(0, 0xFFFFFF)
plt.hist(df['Net Sales Value'],bins=80,color=random_color)
plt.show()
```

```python
sns.histplot(data=df,bins=50)
plt.title('sales_data_with_discounts')
plt.xlabel('interval')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box Plot
sns.boxplot(data=df)
plt.title('sales_data_with_discounts')
plt.xlabel('Variable')
plt.ylabel('Value')
plt.show()
```

```python
sns.barplot(data=df,color='skyblue')
plt.title('sales_data_with_discounts')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```
